# 01 — EDA & Preprocessing

Exploring the 20 Newsgroups dataset and understanding what cleaning we need before feeding text into models.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

from src.utils.data_loader import load_20newsgroups
from src.preprocessing.text_cleaner import TextCleaner

pd.set_option('display.max_colwidth', 120)
print('imports done')
```

```python
# load dataset
train_texts, train_labels, test_texts, test_labels = load_20newsgroups()
print(f'Train: {len(train_texts)} | Test: {len(test_texts)}')
print(f'Classes: {sorted(set(train_labels))}')
```

```python
# class distribution
label_counts = Counter(train_labels)
df_dist = pd.DataFrame(label_counts.items(), columns=['category', 'count']).sort_values('count', ascending=False)

plt.figure(figsize=(12, 5))
sns.barplot(data=df_dist, x='category', y='count', palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Class Distribution — 20 Newsgroups (train set)')
plt.tight_layout()
plt.show()
df_dist
```

```python
# document length distribution
lengths = [len(t.split()) for t in train_texts]
print(f'Min words: {min(lengths)}')
print(f'Max words: {max(lengths)}')
print(f'Mean words: {np.mean(lengths):.1f}')
print(f'Median words: {np.median(lengths):.1f}')

plt.figure(figsize=(10, 4))
plt.hist(lengths, bins=50, edgecolor='black', color='steelblue')
plt.xlabel('Number of words')
plt.ylabel('Frequency')
plt.title('Document Length Distribution')
plt.axvline(np.mean(lengths), color='red', linestyle='--', label=f'Mean = {np.mean(lengths):.0f}')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# look at a raw sample
idx = 0
print(f'Label: {train_labels[idx]}')
print(f'Raw text (first 500 chars):')
print(train_texts[idx][:500])
```

```python
# apply cleaning and compare
cleaner = TextCleaner(remove_stopwords=True, lemmatize=True)
cleaned = cleaner.clean(train_texts[idx])

print('=== ORIGINAL ===')
print(train_texts[idx][:300])
print('\n=== CLEANED ===')
print(cleaned[:300])
```

```python
# most common words per class (after cleaning)
cleaner = TextCleaner()

categories_to_check = ['sci.space', 'talk.politics.misc', 'rec.sport.hockey']

for cat in categories_to_check:
    cat_texts = [t for t, l in zip(train_texts, train_labels) if l == cat]
    cleaned_texts = cleaner.clean_batch(cat_texts[:200])  # sample 200
    all_words = ' '.join(cleaned_texts).split()
    top_words = Counter(all_words).most_common(15)
    print(f'\n[{cat}] Top words:')
    print(', '.join([f'{w}({c})' for w, c in top_words]))
```
